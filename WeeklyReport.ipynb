```python
#South Africa weekly report
import pandas as pd
import datetime as dt

#Obtain current date and week number
today = pd.to_datetime(dt.date.today())
print ("Today is {}".format(str(today)))
thisWeek = today.week
print ("Today's week number is {}.".format(str(thisWeek)))
```

```python
#In the Teams Planner, click on "go to website"`. in the website, click the (...) and "Export plan to excel".
#This excel file path is the imported into this notebook.

df = pd.read_excel (r'C:\Users\Manuel.badillo\Downloads\SA EAG (10).xlsx')

#df.head()
#delete the first 3 rows
new_header = df.iloc[3] #grab the first row for the header
df = df[4:] #take the data less the header row
df.columns = new_header #set the header row as the df header

# Creation of Report dataframe
report = df[['Task Name', 'Progress', 'Assigned To', 'Due Date', 'Late','Description','Bucket Name']].copy()

#convert date format to pandas
report['Due Date'] = pd.to_datetime(report['Due Date'], yearfirst=False, dayfirst=False)

# Remove rows with 'Due Date' = to NaT
report=report.dropna(subset=['Due Date'])

# Add week number column
report['week'] = report['Due Date'].dt.strftime('%V')

# convert week to number
report['week'] = report.week.astype(int)

# Filter to remove not completed tasks and show near deadlines
report = report.loc[(report['Progress'] != "Completed") & (report['week'] == thisWeek)]
report.sort_values(by=['Bucket Name','week'], inplace=True)

#define the function to color red late tasks
def color_negative_red(val):
    """
    Takes a scalar and returns a string with
    the css property `'color: red'` for negative
    strings, black otherwise.
    """
    color = 'red' if val == "true" else 'black'
    return 'color: %s' % color
report = report.style.applymap(color_negative_red)
report
```
